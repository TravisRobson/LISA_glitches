```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
from mpl_toolkits.mplot3d import Axes3D


import sys
sys.path.append("../") 

import LISA as l
```

```python
orb = l.Orbit(l.YEAR)
```

```python
f = np.logspace(-5,-2, 200)
SnX = np.zeros(len(f))

for i in range(0,len(f)):
    SnX[i] = orb.get_Sn(f[i])
```

```python
fig, ax = plt.subplots(1, figsize=(8,6))
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
ax.set_xlabel(r'f [Hz]', fontsize=16)
ax.set_ylabel(r'$\sqrt{S_{n}}~$ [1/Hz]', fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=16)

ax.loglog(f, np.sqrt(SnX))

plt.show()
```

```python
orb.calc_x(1, 100.)
```

```python
t = np.linspace(0, l.YEAR, 100)
x1 = np.zeros(len(t))
y1 = np.zeros(len(t))
z1 = np.zeros(len(t))

x2 = np.zeros(len(t))
y2 = np.zeros(len(t))
z2 = np.zeros(len(t))

x3 = np.zeros(len(t))
y3 = np.zeros(len(t))
z3 = np.zeros(len(t))

div_AU = 1./l.AU

for i in range(0, len(t)):
    orb.calc_x(1, t[i])
    x1[i] = orb.x[0][0]*div_AU
    y1[i] = orb.x[0][1]*div_AU
    z1[i] = orb.x[0][2]*div_AU
    
    orb.calc_x(2, t[i])
    x2[i] = orb.x[1][0]*div_AU
    y2[i] = orb.x[1][1]*div_AU
    z2[i] = orb.x[1][2]*div_AU
    
    orb.calc_x(3, t[i])
    x3[i] = orb.x[2][0]*div_AU
    y3[i] = orb.x[2][1]*div_AU
    z3[i] = orb.x[2][2]*div_AU
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

plt.rc('text', usetex=True)
plt.rc('font', family='serif')
ax.set_xlabel(r'x [AU]', fontsize=16)
ax.set_ylabel(r'y [AU]', fontsize=16)
ax.set_zlabel(r'z [AU]', fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=16)

# the sun
ax.scatter(0.0, 0.0, 0.0, '.', color='orange', s=60)



ax.plot(x1, y1, z1, '--', color='k')
ax.plot(x2, y2, z2, '-.', color='k')
ax.plot(x3, y3, z3, ':', color='k')

# ax.view_init(elev=30, azim=155.)
ax.set_xlim(-1.0, 1.0)
ax.set_zlim(-1.0, 1.0)
ax.set_zlim(-1.0, 1.0)

ax.xaxis.labelpad = 13
ax.yaxis.labelpad = 13
ax.zaxis.labelpad = 11

ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))


# trying to depict the array of lasers
idx=0
a = np.array((x1[idx], x2[idx]))
b = np.array((y1[idx], y2[idx]))
c = np.array((z1[idx], z2[idx]))
ax.plot(a,b,c, '-', linewidth=2, color='r')

a = np.array((x3[idx], x2[idx]))
b = np.array((y3[idx], y2[idx]))
c = np.array((z3[idx], z2[idx]))
ax.plot(a,b,c, '-', linewidth=2, color='r')

ax.scatter(a[1], b[1], c[1], '.', color='k', s=40)

a = np.array((x1[idx], x3[idx]))
b = np.array((y1[idx], y3[idx]))
c = np.array((z1[idx], z3[idx]))
ax.plot(a,b,c, '-', linewidth=2, color='r')

ax.scatter(a[0], b[0], c[0], '.', color='k', s=40)
ax.scatter(a[1], b[1], c[1], '.', color='k', s=40)

xEarth = np.cos(2.*np.pi*t[0]/l.YEAR + 30.*np.pi/180.)
yEarth = np.sin(2.*np.pi*t[0]/l.YEAR + 30.*np.pi/180.)
zEarth = 0.0

ax.scatter(xEarth, yEarth, zEarth, '.', color='blue', s=40)

ax.grid(False)
ax.axis('off')



#plt.savefig('huh.pdf')

plt.show()
```

```python
# merge with executed in terminal
# $convert -delay 20 figs/lisa_orbit*.png LISA_movie.gif
# be sure to set eccentricity to 0.1 if you want a nice graphic!!!

orb = l.Orbit(l.YEAR, ecc=0.1)

# generate the exaggerated data
t = np.linspace(0, l.YEAR, 100)
x1 = np.zeros(len(t))
y1 = np.zeros(len(t))
z1 = np.zeros(len(t))

x2 = np.zeros(len(t))
y2 = np.zeros(len(t))
z2 = np.zeros(len(t))

x3 = np.zeros(len(t))
y3 = np.zeros(len(t))
z3 = np.zeros(len(t))

div_AU = 1./l.AU

for i in range(0, len(t)):
    orb.calc_x(1, t[i])
    x1[i] = orb.x[0][0]*div_AU
    y1[i] = orb.x[0][1]*div_AU
    z1[i] = orb.x[0][2]*div_AU
    
    orb.calc_x(2, t[i])
    x2[i] = orb.x[1][0]*div_AU
    y2[i] = orb.x[1][1]*div_AU
    z2[i] = orb.x[1][2]*div_AU
    
    orb.calc_x(3, t[i])
    x3[i] = orb.x[2][0]*div_AU
    y3[i] = orb.x[2][1]*div_AU
    z3[i] = orb.x[2][2]*div_AU


for i in range(0, len(t)):
    
    fig = plt.figure(figsize=(8,6))
    ax = fig.add_subplot(111, projection='3d')

    plt.rc('text', usetex=True)
    plt.rc('font', family='serif')
    ax.set_xlabel(r'x [AU]', fontsize=16)
    ax.set_ylabel(r'y [AU]', fontsize=16)
    ax.set_zlabel(r'z [AU]', fontsize=16)
    ax.tick_params(axis='both', which='major', labelsize=16)

    ax.plot(x1, y1, z1, '--', color='k')
    ax.plot(x2, y2, z2, '-.', color='k')
    ax.plot(x3, y3, z3, ':', color='k')

    # ax.view_init(elev=30, azim=155.)
    ax.set_xlim(-1.0, 1.0)
    ax.set_zlim(-1.0, 1.0)
    ax.set_zlim(-1.0, 1.0)

    ax.xaxis.labelpad = 13
    ax.yaxis.labelpad = 13
    ax.zaxis.labelpad = 11

    ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
    ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
    ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))


    # trying to depict the array of lasers
    idx=i
    a = np.array((x1[idx], x2[idx]))
    b = np.array((y1[idx], y2[idx]))
    c = np.array((z1[idx], z2[idx]))
    ax.plot(a,b,c, '-', linewidth=2, color='r')

    a = np.array((x3[idx], x2[idx]))
    b = np.array((y3[idx], y2[idx]))
    c = np.array((z3[idx], z2[idx]))
    ax.plot(a,b,c, '-', linewidth=2, color='r')

    ax.scatter(a[1], b[1], c[1], '.', color='k', s=40)

    a = np.array((x1[idx], x3[idx]))
    b = np.array((y1[idx], y3[idx]))
    c = np.array((z1[idx], z3[idx]))
    ax.plot(a,b,c, '-', linewidth=2, color='r')

    ax.scatter(a[0], b[0], c[0], '.', color='k', s=40)
    ax.scatter(a[1], b[1], c[1], '.', color='k', s=40)
    
    # the sun
    ax.scatter(0.0, 0.0, 0.0, '.', color='orange', s=60)
    
    # track the earth too!
    xEarth = np.cos(2.*np.pi*t[i]/l.YEAR + 30.*np.pi/180.)
    yEarth = np.sin(2.*np.pi*t[i]/l.YEAR + 30.*np.pi/180.)
    zEarth = 0.0

    ax.scatter(xEarth, yEarth, zEarth, '.', color='blue', s=40)

    ax.grid(False)
    ax.axis('off')
    
    title = '../Animations/figs/lisa_orbit_'
    title = title + str(i).zfill(3) + '.png'
    
    
    
    plt.savefig(title, dpi=300)
    
```
